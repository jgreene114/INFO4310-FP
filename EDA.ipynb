```python

result = df.groupby('SID').agg({'ISO_TIME': lambda x: x.dt.year.mode()[0], 'NAME': lambda x: x.mode()[0], 'USA_SSHS': 'max'}).reset_index()
```

```python
counts = result.groupby(['ISO_TIME', 'USA_SSHS']).size().reset_index(name='Count')
```

```python
counts = counts[counts['USA_SSHS'] >= 1]
```

```python
counts
```

```python
window_size=3
counts['Moving Average'] = counts.groupby('USA_SSHS')['Count'].transform(
    lambda x: x.rolling(window_size, min_periods=1).mean()
)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(data=counts, x='ISO_TIME', y='Moving Average', hue='USA_SSHS', marker='o')
plt.title('Number of Storms by SSHS Value Year by Year')
plt.xlabel('Year')
plt.ylabel('Number of Storms')
plt.legend(title='SSHS Value')
plt.grid(True)
plt.show()
```

```python
result['SSHS_Group'] = pd.cut(result['USA_SSHS'], bins=[-float('inf'), 2, float('inf')], labels=['minor', 'major'])
counts = result.groupby(['ISO_TIME', 'SSHS_Group']).size().reset_index(name='Count')
window_size = 5
counts['Moving Average'] = counts.groupby('SSHS_Group')['Count'].transform(
    lambda x: x.rolling(window_size, min_periods=1).mean()
)
```

```python
counts
```

```python
plt.figure(figsize=(10, 6))

counts2 = counts[counts['ISO_TIME'] >= 1980]

sns.lineplot(data=counts2, x='ISO_TIME', y='Moving Average', hue='SSHS_Group', marker='o')

for group, data in counts2.groupby('SSHS_Group'):
    sns.regplot(x='ISO_TIME', y='Count', data=data, ci=None, scatter=False, label=f'Trendline {group}')



plt.title('Moving Average of Number of Storms by SSHS Group Year by Year')
plt.xlabel('Year')
plt.ylabel('Moving Average of Number of Storms')
plt.legend(title='SSHS Group')
plt.grid(True)
plt.show()
```
